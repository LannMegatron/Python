```python
# TUGAS
# Bilangan Prima (Minimal 5 Bilangan)
# Type String (Minimal 5 Element)
# Gunakan Perintah: Index(Minimal 3 Element), Slicing(Minimal 3x), Gabung(Tambah Minimal 3 Element) Untuk SOAL DI ATAS!

# Index
list = [3, 11, 7, 5, 2]

print('index list')
print(list[0])
print(list[2])
print(list[4])
print(list[-1])
print(list[-2])
print(list[-5])
print('banyaknya element')
print(len(list))

# Slicing
print('sebelum dipotong')
print(list)
print('setelah dipotong')
print(list[-1:])
print(list[:-1])
print(list[:2])

# Gabung
print('sebelum digabung')
print(list)
print('setelah digabung')
print(list + [13, 19, 17])

# Index
list = ['Rusia', 'Germany', 'Indonesia', 'Portugal', 'Argentina']
print('index list')
print(list[0])
print(list[2])
print(list[4])
print(list[-1])
print(list[-2])
print(list[-5])
print('banyaknya element')
print(len(list))

# Slicing
print('sebelum dipotong')
print(list)
print('setelah dipotong')
print(list[-1:])
print(list[:-1])
print(list[:2])

# Gabung
print('sebelum digabung')
print(list)
print('setelah digabung')
print(list + ['Japan', 'Korea', 'China'])
```
